```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
```

```python
 iris=load_iris()
```

```python
iris.DESCR
```

```python
print('Features:', iris.feature_names)
print('Target:', iris.target_names)
```

```python
print('data:', iris.data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,  y_test = train_test_split(iris.data, iris.target, random_state=10, test_size=0.20)
```

```python
DTC=DecisionTreeClassifier(max_depth=2, random_state=0)
```

```python
DTC.fit(X_train, y_train)
```

```python
y_pred= DTC.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix (y_test, y_pred)
cm
```

```python
tree.plot_tree(DTC)
```

```python
from sklearn.metrics import classification_report
print('clr', classification_report(y_test, y_pred))
```

```python
RFC=RandomForestClassifier(n_estimators=50)
RFC.fit(X_train, y_train)
y_pred=RFC.predict(X_test)
print('CLF', classification_report(y_test, y_pred))
```

```python
fn=iris.feature_names
cn=iris.target_names
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=800)
tree.plot_tree(RFC.estimators_[0],
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('rf_individualtree.png')
```
